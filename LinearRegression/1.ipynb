```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("C:/Users/Sathvik K/OneDrive/Desktop/Book1.csv")
df
```

```python
%matplotlib inline
f,a = plt.subplots()
a.set_xlabel("area",fontsize = 20)
a.set_ylabel("price",fontsize = 20)
plt.scatter(df.area,df.price, color = 'purple', marker = '1')
```

```python
reg = linear_model.LinearRegression().fit(df[['area']],df.price)

plt.scatter(df.area,df.price,color = 'purple', marker = '.')
plt.plot(df.area,reg.predict(df[['area']]),color = 'blue')
```
